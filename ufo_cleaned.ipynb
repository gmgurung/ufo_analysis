```python
def get_int1(data):
    int_val = []
    for item in data:
        clean_dtr = ''.join(filter(str.isdigit,item))
        if clean_dtr:
            int_val.append(int(clean_dtr))
    return int_val
```

```python
get_int1(df1["duration (hours/min)"])
number = pd.Series(get_int1(df1["duration (hours/min)"]))
```
